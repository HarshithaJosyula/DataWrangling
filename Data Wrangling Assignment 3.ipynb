```python
import pandas as pd
```

```python
df = pd.read_csv('DW3_set_exercise.csv')
```

```python
df.head()
```

```python
#1
diabetes_codes = ['E08', 'E09', 'E10', 'E11', 'E13']
diabetes_patients = df[df['Diagnosis Code'].str.startswith(tuple(diabetes_codes))]
diabetes_patient_ids = diabetes_patients['Patient ID'].tolist()
print("Patients with Diabetes IDs:", diabetes_patient_ids)

cardinality_diabetes_set = len(diabetes_patient_ids)
print("Cardinality of the Diabetes set:", cardinality_diabetes_set)
```

```python
#2

covid_codes = ['U07.1', 'J12.82']
covid_patients = df[df['Diagnosis Code'].str.startswith(tuple(covid_codes))]
covid_patient_ids = covid_patients['Patient ID'].tolist()
print("Patients with COVID IDs:", covid_patient_ids)

cardinality_covid_set = len(covid_patient_ids)
print("Cardinality of the COVID set:", cardinality_covid_set)
```

```python
# Maybe Pneumonia due to COVID-19 code patient ids are already in the covid code
covid_patient_ids_set = set(covid_patient_ids)

if len(covid_patient_ids) != len(covid_patient_ids_set):
    print("There are duplicates in the list of COVID patient IDs.")
else:
    print("There are no duplicates in the list of COVID patient IDs.")
```

```python
covid_patient_ids_set = set(covid_patient_ids)

covid_patient_ids_unique = list(covid_patient_ids_set)

print("Unique COVID patient IDs:", covid_patient_ids_unique)
print("cardinality: ", len(covid_patient_ids_unique))
```

```python
# 3
intersection_patient_ids = list(set(diabetes_patient_ids) & set(covid_patient_ids))

print("Patients with both Diabetes and COVID IDs:", intersection_patient_ids)

cardinality_intersection_set = len(intersection_patient_ids)
print("Cardinality of the Intersection set:", cardinality_intersection_set)
```

```python
#4

union_patient_ids = list(set(diabetes_patient_ids) | set(covid_patient_ids))
print("Patients with either Diabetes or COVID IDs:", union_patient_ids)

cardinality_union_set = len(union_patient_ids)
print("Cardinality of the Union set:", cardinality_union_set)
```

```python
#5
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted


sizes = {
    '10': len(diabetes_patient_ids) - cardinality_intersection_set,  # Diabetes only
    '01': len(covid_patient_ids) - cardinality_intersection_set,  # COVID only
    '11': cardinality_intersection_set,  # Intersection
}


venn2(subsets=sizes, set_labels=('Diabetes', 'COVID'))


plt.title("Venn Diagram of Diabetes and COVID Sets")

plt.text(0.8, 0.3, str(cardinality_union_set), fontsize=12, color='blue')
plt.text(0.6, 0.8, str(cardinality_union_set), fontsize=12, color='orange')
plt.text(0.4, 0.3, str(len(diabetes_patient_ids)), fontsize=12, color='blue')
plt.text(0.6, 0.1, str(len(covid_patient_ids)), fontsize=12, color='orange')
plt.text(0.5, 0.5, str(cardinality_intersection_set), fontsize=12, color='green')


venn2_circles(subsets=sizes, linestyle='solid')
plt.show()
```

```python
#6
covid_patients = df[df['Diagnosis Code'].str.startswith(tuple(covid_codes))]
covid_patient_dates = covid_patients[['Patient ID', 'Diagnosis Code', 'Date']]
diabetes_patients = df[df['Diagnosis Code'].str.startswith(tuple(diabetes_codes))]
diabetes_patient_dates = diabetes_patients[['Patient ID', 'Diagnosis Code', 'Date']]

merged_data = pd.merge(covid_patient_dates, diabetes_patient_dates, on='Patient ID', suffixes=('_covid', '_diabetes'))

merged_data['Date_covid'] = pd.to_datetime(merged_data['Date_covid'])
merged_data['Date_diabetes'] = pd.to_datetime(merged_data['Date_diabetes'])

# Filter patients where diabetes diagnosis occurred after COVID diagnosis
diabetes_only_after_covid = merged_data[merged_data['Date_diabetes'] > merged_data['Date_covid']]


patients_diabetes_only_after_covid = diabetes_only_after_covid['Patient ID'].unique().tolist()
print("Patients with Diabetes only after COVID IDs:", patients_diabetes_only_after_covid)

cardinality_diabetes_only_after_covid = len(patients_diabetes_only_after_covid)
print("Cardinality of the Diabetes only after COVID set:", cardinality_diabetes_only_after_covid)

print("Count breakdown for each diabetes code occurring only after COVID:")
diabetes_only_after_covid_count = diabetes_only_after_covid['Diagnosis Code_diabetes'].value_counts()
print(diabetes_only_after_covid_count)
```
